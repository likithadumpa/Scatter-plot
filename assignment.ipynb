```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Assignment_dataset.csv',parse_dates=['Date'])
```

```python
x1=data[data['GHI']<2]
```

```python
x2 = data[(data['GHI']>=2) & (data['GHI']<4)]
```

```python
x3=data[(data['GHI']>=4)&(data['GHI']<6)]
```

```python
x4 = data[data['GHI']>=6]
```

```python
plt.scatter(x1['Date'],x1['PR'],marker="D",c="#000080",s=10)
plt.scatter(x2['Date'],x2['PR'],marker="D",c="#ADD8E6",s=10)
plt.scatter(x3['Date'],x3['PR'],marker="D",c="orange",s=10) 
plt.scatter(x4['Date'],x4['PR'],marker="D",c="brown",s=10) 
plt.xlabel('Daily Irradiation[kWh/m2]') 
plt.ylabel('Performance Ratio')
plt.show()
```

```python
pr = data['PR']
```

```python
sum=0

for i in range(0,982):
    sum = sum+pr[i]
    avg = sum/(i+1)
    pr[i] = avg 
```

```python
plt.scatter(x1['Date'],x1['PR'],marker="D",c="#000080",s=10)
plt.scatter(x2['Date'],x2['PR'],marker="D",c="#ADD8E6",s=10)
plt.scatter(x3['Date'],x3['PR'],marker="D",c="orange",s=10)
plt.scatter(x4['Date'],x4['PR'],marker="D",c="brown",s=10)
plt.xlabel('Daily Irradiation[kWh/m2]')
plt.ylabel('Performance Ratio')
plt.plot(data['Date'],pr,c="red")
plt.show()
```
